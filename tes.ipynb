```python
from tensorflow.keras.models import load_model

# Memuat seluruh model
model = load_model('model_cats_dogs.h5')
```

```python
import numpy as np 
from keras.preprocessing import image

path = 'D:/Belajar Data Science/UDEMY/CNN_Cats-Dogs/CNN_File/Section 40 - Convolutional Neural Networks (CNN)/dataset/single_prediction/cat_or_dog_9.jpg'
test_image = image.load_img(path, target_size = (64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
result = model.predict(test_image/255.0)
if result[0][0] > 0.5:
    prediction = 'dog'
else:
    prediction = 'cat'
    


import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Membaca gambar
img = mpimg.imread(path)

# Menampilkan gambar
plt.imshow(img)
print("image classified as a", prediction)
```
